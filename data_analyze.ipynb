```python
def aggregate_infos(isu):
    general_info = general_infos[general_infos['isu'] == isu]
    publication = publications[publications['isu'] == isu]
    project = projects[projects['isu'] == isu]
    event = events[events['isu'] == isu]
    
    return {
        'general_info': general_info.reset_index(drop=True),
        'publication': publication.reset_index(drop=True),
        'project': project.reset_index(drop=True),
        'event': event.head().reset_index(drop=True)
    }
```

```python
def save_aggregate_infos(isu, data):
    with pd.ExcelWriter(f'aggregate_info/aggregate_info_{isu}.xlsx') as writer:
        data['general_info'].to_excel(writer, sheet_name='general_infos')
        data['publication'].to_excel(writer, sheet_name='publications')
        data['project'].to_excel(writer, sheet_name='projects')
        data['event'].to_excel(writer, sheet_name='events')
```

```python
for index, row in general_infos.iterrows():
    data = aggregate_infos(row['isu'])
    save_aggregate_infos(row['isu'], data)
```
